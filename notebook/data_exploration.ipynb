### Exploring oil price data

```python
data = pd.read_csv("../data/oil.csv")
```

```python
plt.scatter(data['date'], data['dcoilwtico'], c ="blue")
plt.show()
```

```python
data = data.dropna(axis=0)
```

```python
data.count()
```

```python
plt.scatter(data['date'], data['dcoilwtico'], c ="blue")
plt.show()
```

```python
data = data.sort_values(by=['date'])
```

```python
mx_dt = data.iloc[-1].date
mn_dt = data.iloc[0].date
dt_delta = datetime.timedelta(days = 10)
mn_dt = datetime.datetime.strptime(mn_dt,'%Y-%m-%d')
mx_dt = datetime.datetime.strptime(mx_dt,'%Y-%m-%d')
grouped_dt = drange(mn_dt, mx_dt, dt_delta)
y = np.arange(len(grouped_dt))
```

```python
plt.plot_date(grouped_dt, y ** 2)
plt.title('matplotlib.pyplot.plot_date() function Example', fontweight ="bold")
plt.show()
```

```python
plt.plot_date(data['date'], data['dcoilwtico'])
plt.title('matplotlib.pyplot.plot_date() function Example', fontweight ="bold")
plt.show()
```

### Exploring Transaction data

```python
trans_dta = pd.read_csv("../data/transactions.csv")
```

```python
trans_dta = trans_dta.sort_values(['date'])
```

```python
trans_dta.head()
```

```python
print(trans_dta[(trans_dta['date']=='2013-01-02') & (trans_dta['store_nbr']==31)].transactions.empty)
```

```python
trans_dta.iloc[-1]['date']
```

```python
FORMAT = "%Y-%m-%d"
start = datetime.strptime(trans_dta.iloc[0]['date'], FORMAT)
end = datetime.strptime(trans_dta.iloc[-1]['date'], FORMAT)
delta = dt.timedelta(days=int("1"))
curr = start
dt_list = []
while curr <= end:
    dt_list.append(curr.strftime(FORMAT))
    curr += delta

trans = {}
for i in range(1,55):
    trans[i] = []
    
for idt in dt_list[:10]:
    for i in range(1,2):
        value = 0 if trans_dta[(trans_dta['date']==idt) & (trans_dta['store_nbr']==i)].empty else trans_dta[(trans_dta['date']==idt) & (trans_dta['store_nbr']==i)].transactions
        trans[i].append(value)
        
print(trans[1])
```
